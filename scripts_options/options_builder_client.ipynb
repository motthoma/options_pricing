# Simple Long Call

```python
builder = OptionBuilder()

k = 50  # Strike price for long put
p = 8   # Premium for long put

# Build the option strategy (Long using Calls)
option_strategy = (
    builder.build_long_call(k, p)
            .build()
)

stock_price = 45
result = option_strategy.evaluate(stock_price)
print(f"Option strategy result at stock price {stock_price}: {result}")

# Plot the strategy over a range of stock prices
option_strategy.plot(stock_price_range=(0, 60), strategy_name="Long Call")
```

# Bull Spread with calls

```python
builder = OptionBuilder()

# Define the strikes and premiums
k_long_call = 25  # Strike price for long put
p_long_call = 10   # Premium for long put

k_short_call = 30  # Strike price for short put
p_short_call = 8   # Premium for short put


# Build the option strategy
option_strategy = (
    builder.build_long_call(k_long_call, p_long_call)
           .build_short_call(k_short_call, p_short_call)
           .build()
)

# Evaluate the strategy at a given stock price
stock_price = 30
result = option_strategy.evaluate(stock_price)
print(f"Option strategy result at stock price {stock_price}: {result}")

# Plot the strategy over a range of stock prices
option_strategy.plot(stock_price_range=(0, 60), strategy_name="Bull Spread Calls")
```

# Butterfly Spread with Puts

```python
builder = OptionBuilder()

# Define the strikes and premiums
k_long_put_1 = 25  # Strike price for long put
p_long_put_1 = 3   # Premium for long put

k_short_put = 30  # Strike price for short put
p_short_put = 3   # Premium for short put

k_long_put_2 = 35  # Strike price for another long put
p_long_put_2 = 3   # Premium for the second long put

# Build the option strategy (butterfly spread using puts)
option_strategy = (
    builder.build_long_put(k_long_put_1, p_long_put_1)
            .build_short_put(k_short_put, p_short_put)
            .build_short_put(k_short_put, p_short_put)
            .build_long_put(k_long_put_2, p_long_put_2)
            .build()
)

# Evaluate the strategy at a given stock price
stock_price = 30
result = option_strategy.evaluate(stock_price)
print(f"Option strategy result at stock price {stock_price}: {result}")

# Plot the strategy over a range of stock prices
option_strategy.plot(stock_price_range=(0, 60), strategy_name="Butterfly_Puts")
```
